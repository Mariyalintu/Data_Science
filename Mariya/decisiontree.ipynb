```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data=datasets.load_iris()
print(data.DESCR)
print(data.target)
print(data.feature_names)
```

```python
print(data.target_names)
```

```python
x=data.data
print(x)
```

```python
y=data.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=1)
dt=DecisionTreeClassifier(criterion="entropy",splitter="best",max_depth=None,min_samples_leaf=1)
dt.fit(x_train,y_train)
y_predict=dt.predict(x_test)
```

```python
print("------test Class ------\n",y_test)
print("------Predict Class ------\n",y_predict)
```

```python
f=plt.figure(figsize=(10,10))
tree.plot_tree(dt)
plt.show(f)
```

```python
cm=confusion_matrix(y_test,y_predict,normalize=None,labels=dt.classes_)
```

```python
cmdis = ConfusionMatrixDisplay(confusion_matrix=cm,)
cmdis.plot()
```
