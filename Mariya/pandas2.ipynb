```python
print("data Cleaning\n")
df=pd.read_csv('food_cleaning.csv')
print(df)
```

```python
df.duplicated().any()
df.drop_duplicates(inplace=True)
print(df)
```

```python
mean_sweetness=df['Crunchiness'].mean()
print(mean_sweetness)
df['Sweetness'].fillna(mean_sweetness,inplace=True)
print(df)
```

```python
df['Sweetness'] = pd.to_numeric(df['Sweetness'],errors='coerce')
print(df)
```

```python
df = df[df['FoodType'].isin(['fruit', 'vegetable', 'protein'])]
df = df.dropna()
print(df)
```

```python
mean_values = df.groupby('FoodType').mean()
print(mean_values)
```

```python
for x in df.index:
    if df.loc[x,'FoodType'] != 'fruit' and df.loc[x,'FoodType'] != 'vegetable':
        df.drop(x,inplace=True)
print(df)
```
