```python
import numpy as np
from scipy import ndimage
from PIL import Image
from matplotlib import pyplot as plt
a=Image.open('m1.jpg')
b=np.array(a)

```

```python
plt.imshow(b)
```

```python
x=ndimage.rotate(b,270)
d=Image.fromarray(x)
print(d)
plt.imshow(d)
```

```python
z=ndimage.zoom(b,(.25,.25,1))
z2=Image.fromarray(z)
plt.imshow(z2)
```

```python
p=np.array([[1,0,0],[0,1,0],[0,0,1]])
x1=ndimage.affine_transform(b,p)
y=Image.fromarray(x1)
plt.imshow(y)
```
