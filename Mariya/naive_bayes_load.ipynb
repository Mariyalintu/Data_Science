```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
a=iris.data
print(a)
```

```python
b=iris.target
print(b)
```

```python
print(iris.DESCR)
```

```python
print(iris.feature_names)
```

```python
print(iris.target_names)
```

```python
a_train,a_test,b_train,b_test=train_test_split(a,b,train_size=.75,random_state=5)
```

```python
nb=GaussianNB()
nb.fit(a_train,b_train)
b_predict=nb.predict(a_test)
print("------Test  Class --------\n",b_test)
print("-----Predict Class -------\n",b_predict)
```

```python
print(accuracy_score(b_test,b_predict,normalize=True))
```

```python
print(accuracy_score(b_test,b_predict,normalize=False))
```

```python
cm=confusion_matrix(b_test,b_predict,normalize=None)
print(cm)
```

```python
cm_dis=ConfusionMatrixDisplay(cm)
cm_dis.plot()
plt.show
```
